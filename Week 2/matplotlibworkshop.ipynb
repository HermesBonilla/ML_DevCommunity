````
import matplotlib.pyplot as plt
import matplotlib.image as img 
import numpy as np
import pandas as pd 
# Graphs as cell output
%matplotlib inline 

# We can display images, but it's not the main use of matplot
dscImage = img.imread("DSC_Logo.png")
plt.imshow(dscImage)
````

````
# Created arbitrary data
# age and occupation salary

ages_x = [21,25,28,29,33,34,35,39,40,44]
py_salary_y = [38000, 42500, 54000, 54300, 71800,
        73300, 75000, 85430, 87470, 95550] 
````

````
plt.plot(ages_x, py_salary_y) # x, y
````

````
# Plotting Python developers only
plt.plot(ages_x, py_salary_y, label='Python') 
plt.xlabel('Ages')
plt.ylabel('Average Salary ($)')
plt.title('Average Salary of Developers by Age')
plt.legend(['Python'])

plt.show()
````

````
# Introducing new arbitrary data on java salaries

java_salary_y = [36000, 44000, 54000, 54300, 80500,
                81000, 83000, 85430, 87470, 90000] 
````

````
# Plotting both Java & Python developer salaries
plt.plot(ages_x,py_salary_y, color='#3776ab', label='Python')
plt.plot(ages_x, java_salary_y, color='#FF6700', label='Java')
plt.xlabel('Ages')
plt.ylabel('Average Salary ($)')
plt.title('Average Salary of Developers by Age')
plt.legend(['Python', 'Java'])

plt.show()
````

````
# Graphing categorical values

names = ['Yabba', 'Dabba', 'Doo'] # Names of our categorical values
values = [1, 10, 100]  # The hypothetical amount of each

plt.figure(figsize=(9, 3)) # Adjusting the size of the main figure

# Here we are creating 3 subplots inside of the main figure each a different type of graph
plt.subplot(132) 
plt.bar(names, values)
plt.subplot(133)
plt.scatter(names, values)
plt.subplot(131)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

````
#Working with text, adding labels to the graphs

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, 50, density=1, facecolor='b', alpha=0.75)


plt.xlabel('Smarts')      # Used to add label for X axis
plt.ylabel('Probability') # Used to add label for Y axis
plt.title('Histogram of IQ') # Used to add title for graph
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # Used to add test inside the graph, the TeX is for the symbols
                            # TeX is string notation for LaTeX the MS Word of math which is built in to matplotlib

plt.axis([40, 160, 0, 0.03]) # Used to change the values shown in the axis x = (40,160), y = (0, 0.03)
plt.grid(True) #Used to show gridlines in the graph

plt.show()
````

## Importing our real world data

````
df1 = pd.read_csv('/work/2019 - 01.csv')
df2 = pd.read_csv('/work/2019 - 02.csv')
df3 = pd.read_csv('/work/2019 - 03.csv')
df4 = pd.read_csv('/work/2019 - 04.csv')
df5 = pd.read_csv('/work/2019 - 05.csv')
````

## Let's combine our data

Pandas documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html

````
#creating the mega dataframe

frames = [df1, df2, df3, df4, df5]
dfmega = pd.concat(frames)
````

````
dfmega.head()
````

````
dfmega
````

````
print("dfmega = ",pd.unique(dfmega['month']))
````

````
#Rearrange our data

cols = dfmega.columns.tolist()
print(cols)
````

````
cols = [cols[0]] + cols[2:] + [cols[1]]
print(cols)
````

````
dfmega = dfmega[cols]
````

````
dfmega.head()
````

````
dfmega.drop(columns=['start_station_id','end_station_id','bike_id'],inplace=True)
````

````
dfmega.head()
````

## Visualizing our data

Relevant Tasks:
- Find which months have the longest trips on average/median
- Find which gender has the longest trips on average/median
- Find which age group have the longest trips on average/median
- Find which type of user has the longest trips on average/median
- Find relationships between the start locations and end locations

````
dfmega.groupby(['month']).mean()
````

````
monthtrip = dfmega.groupby(['month']).mean()
monthtrip.drop('member_birth_year',inplace=True,axis=1)
monthtrip
````

````
monthtrip.plot.bar();
````

````
dfmega.groupby(['month']).median()
````

````
monthtrip = dfmega.groupby(['month']).median()
monthtrip.drop('member_birth_year',inplace=True,axis=1)
monthtrip
````

````
monthtrip.plot.barh();
````

````
dfmega.groupby(['member_gender']).mean()
````

````
gentrip = dfmega.groupby(['member_gender']).mean()
gentrip.drop('member_birth_year',inplace=True,axis=1)
gentrip
````

````
gentrip.plot.pie(subplots=True,figsize=(6, 6),legend=None);
````

````
dfmega.groupby(['member_gender']).median()
````

````
gentrip = dfmega.groupby(['member_gender']).median()
gentrip.drop('member_birth_year',inplace=True,axis=1)
gentrip
````

````
gentrip.plot.pie(subplots=True,figsize=(6, 6),legend=None,autopct="%.2f");
````
